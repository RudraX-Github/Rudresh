```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the target URL
url = "https://leg.wa.gov/legislators"  # Replace with the actual URL

try:
    # Send a GET request
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find all records
        records = soup.find_all("li")  # Modify the tag if necessary

        # Open a CSV file for writing
        with open("legislators_data.csv", "w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            
            # Write the header row
            writer.writerow(["ID", "Name", "Chamber", "District", "Party", "Committees", "Leadership"])

            # Loop through records and extract data
            for record in records:
                data_id = record.get("data-id", "N/A")
                data_name = record.get("data-name", "N/A")
                data_chamber = record.get("data-chamber", "N/A")
                data_district = record.get("data-district", "N/A")
                data_party = record.get("data-party", "N/A")
                data_committees = record.get("data-committees", "N/A")
                data_leadership = record.get("data-leadership", "N/A")

                # Write each record to CSV
                writer.writerow([data_id, data_name, data_chamber, data_district, data_party, data_committees, data_leadership])
        
        print("\n✅ Data successfully extracted and saved to legislators_data.csv")

    else:
        print(f"\n❌ Error: HTTP {response.status_code} - Unable to fetch data.")

except requests.exceptions.RequestException as e:
    print(f"\n⚠️ Request failed: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the main page URL
base_url = "https://leg.wa.gov/legislators"  # Replace with the actual website

try:
    # Fetch the main page content
    response = requests.get(base_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        # Find all links (URLs)
        records = soup.find_all("li")  # Modify if necessary

        # Open CSV file for saving data
        with open("legislators_addresses.csv", "w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow(["Office Name", "Address", "Office Number", "Contact Number"])

            # Loop through each record to extract the dynamic URL
            for record in records:
                relative_url = record.find("a")["href"]  # Extract href link
                full_url = f"href="/legislators/member/peter-abbarno"{relative_url}"  # Ensure full URL format

                # Fetch the details page
                detail_response = requests.get(full_url)
                
                if detail_response.status_code == 200:
                    detail_soup = BeautifulSoup(detail_response.text, "html.parser")
                    
                    # Extract address details from the contact-list
                    contact_section = detail_soup.find(class_="contact-list")
                    
                    if contact_section:
                        office_name = contact_section.find(class_="text-lg").text.strip()
                        address = contact_section.find(class_="text-md contact-block-indent").text.strip()
                        office_number = contact_section.find_all(class_="text-lg")[1].text.strip()  # Second occurrence
                        contact_number = contact_section.find_all(class_="text-md contact-block-indent")[1].text.strip()  # Second occurrence

                        # Write extracted data into CSV
                        writer.writerow([office_name, address, office_number, contact_number])

        print("\n✅ Data successfully extracted and saved to legislators_addresses.csv")

    else:
        print(f"\n❌ Error: HTTP {response.status_code} - Unable to fetch main page data.")

except requests.exceptions.RequestException as e:
    print(f"\n⚠️ Request failed: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://leg.wa.gov"  # Base website URL

# Open CSV file for saving extracted data
with open("legislators_addresses.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Legislator Name", "Office Name", "Address", "Office Number", "Contact Number"])

    # Fetch the main page containing legislator listings
    response = requests.get(base_url + "/legislators")  

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find all legislator links
        legislators = soup.find_all("a", class_="title")

        for leg in legislators:
            relative_url = leg["href"]  # Extract relative URL
            full_url = f"{base_url}{relative_url}"  # Construct full URL
            
            legislator_name = leg.get_text(strip=True).replace("Representative", "").replace("Senator", "").strip()

            print(f"Visiting: {full_url}")

            # Fetch legislator's page
            legislator_response = requests.get(full_url)

            if legislator_response.status_code == 200:
                legislator_soup = BeautifulSoup(legislator_response.text, "html.parser")
                
                # Extract office details from the contact section
                contact_section = legislator_soup.find(class_="contact-list")
                
                if contact_section:
                    office_name = contact_section.find(class_="text-lg").text.strip()
                    address = contact_section.find(class_="text-md contact-block-indent").text.strip()
                    office_number = contact_section.find_all(class_="text-lg")[1].text.strip()
                    contact_number = contact_section.find_all(class_="text-md contact-block-indent")[1].text.strip()

                    # Write extracted data into CSV
                    writer.writerow([legislator_name, office_name, address, office_number, contact_number])

        print("\n✅ Data successfully extracted and saved to legislators_addresses.csv")

    else:
        print(f"\n❌ Error loading main page: {response.status_code}")
```
